```python
perf_df = pd.read_csv("../../data/raw/07_scheme_performance.csv")

print("Shape:", perf_df.shape)
perf_df.head()
```

```python
perf_df.info()
```

```python
return_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct"
]

for col in return_cols:
    perf_df[col] = pd.to_numeric(
        perf_df[col],
        errors="coerce"
    )

print(perf_df[return_cols].isnull().sum())
```

```python
anomaly_expense = perf_df[
    (perf_df["expense_ratio_pct"] < 0.1) |
    (perf_df["expense_ratio_pct"] > 2.5)
]

print("Expense Ratio Anomalies:")
print(len(anomaly_expense))
```

```python
print(perf_df.isnull().sum())
```

```python
anomaly_returns = perf_df[
    (perf_df["return_1yr_pct"] > 100) |
    (perf_df["return_3yr_pct"] > 100) |
    (perf_df["return_5yr_pct"] > 100)
]

print("Return Anomalies:", len(anomaly_returns))
```

```python
perf_df.to_csv(
    "../../data/processed/scheme_performance_cleaned.csv",
    index=False
)

print("Scheme Performance Cleaned Successfully!")
```
