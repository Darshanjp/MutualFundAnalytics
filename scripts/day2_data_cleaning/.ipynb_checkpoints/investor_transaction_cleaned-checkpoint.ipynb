```python
import pandas as pd

trans_df = pd.read_csv("../../data/raw/08_investor_transactions.csv")

print(trans_df.columns.tolist())
trans_df.head()
```

```python
trans_df["transaction_type"].unique()
```

```python
trans_df["transaction_type"] = (
    trans_df["transaction_type"]
    .str.strip()
    .str.upper()
)

trans_df["transaction_type"] = trans_df["transaction_type"].replace({
    "LUMPSUM": "Lumpsum",
    "SIP": "SIP",
    "REDEMPTION": "Redemption"
})
```

```python
print(trans_df["transaction_type"].unique())
```

```python
print(trans_df["kyc_status"].unique())

print(trans_df.isnull().sum())
```

## Investor Transactions Data Cleaning

### Validation Results

- Transaction types standardized
- Amount values validated (>0)
- Transaction date converted to datetime
- KYC status validated
- Missing values checked

### Results

- Invalid Amount Records: 0
- Missing Values: 0
- KYC Status Values:
  - Verified
  - Pending

### Output File

data/processed/investor_transactions_cleaned.csv


```python
trans_df.to_csv(
    "../../data/processed/investor_transactions_cleaned.csv",
    index=False
)

print("Investor Transactions Cleaned Successfully!")
```
