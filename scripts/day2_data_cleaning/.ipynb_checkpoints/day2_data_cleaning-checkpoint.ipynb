```python
import pandas as pd
import numpy as np
```

```python
nav_df = pd.read_csv("../data/raw/02_nav_history.csv")

nav_df.head()
```

```python
print("Shape:", nav_df.shape)
```

```python
nav_df["date"] = pd.to_datetime(nav_df["date"])
```

```python
nav_df.info()
```

```python
nav_df = nav_df.sort_values(["amfi_code", "date"])
```

```python
print("Duplicate Rows:", nav_df.duplicated().sum())
```

```python
invalid_nav = nav_df[nav_df["nav"] <= 0]

print("Invalid NAV Records:", len(invalid_nav))
```

```python
print(nav_df.isnull().sum())
```

```python
nav_df["nav"] = nav_df["nav"].ffill()
```

```python
nav_df.to_csv(
    "../data/processed/nav_history_cleaned.csv",
    index=False
)

print("NAV History Cleaned Successfully!")
```
